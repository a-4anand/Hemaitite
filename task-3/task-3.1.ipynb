Write a Python script which will accept an url from user and download the images in a folder.(Also it will show total images and how many images downloaded from the web page)
Ex: 78 images downloaded from 100 images

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
url = "https://lavasa.christuniversity.in/academics/Main-Campus/school-of-sciences" 
```

```python
save_dir = "downloaded_images"
os.makedirs(save_dir, exist_ok=True)
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup.prettify())
```

```python
mg_tags = soup.find_all("img")
```

```python
for img in mg_tags:
    img_url = img.get("src")
    if img_url:
        
        img_url = urljoin(url, img_url)

        img_data = requests.get(img_url).content
        
        img_name = os.path.join(save_dir, img_url.split("/")[-1])
       
        with open(img_name, "wb") as f:
            f.write(img_data)
        print(f"Downloaded: {img_name}")

print("All images downloaded successfully.")
```
